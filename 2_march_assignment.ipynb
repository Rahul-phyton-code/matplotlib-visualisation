2.) A scatter plot is a visual representation of how two variables relate to each other. You can use scatter plots to explore the relationship between two variables, for example by looking for any correlation between them.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
plt.scatter(x,y,c= 'g',alpha = 0.9, marker = 'X')
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
plt.title("Difference Between x vs y")
plt.grid()
```

3.) The matplotlib.pyplot.subplots method provides a way to plot multiple plots on a single figure. Given the number of rows and columns, it returns a tuple (fig, ax), giving a single figure fig with an array of axes ax.



```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(1)
plt.subplot(2,2,1)
plt.title("line1")
plt.plot([0,1,2,3,4,5],[0,100,200,300,400,500])
plt.subplot(2,2,2)
plt.title("line2")
plt.plot([0,1,2,3,4,5],[50,20,40,20,60,70])
plt.subplot(2,2,3)
plt.title("line3")
plt.plot([0,1,2,3,4,5],[10,20,30,40,50,60])
plt.subplot(2,2,4)
plt.title("line4")
plt.plot([0,1,2,3,4,5],[200,350,250,550,450,150])
```

4.) A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.
 

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple","Microsoft","Google","AMD"])
Profit = np.array([3000,8000,1000,10000])
plt.bar(company,Profit,color='b')
plt.title("company vs profit")
plt.xlabel("various company brands")
plt.ylabel("profit")
plt.grid()

```

5.) Boxplots are a measure of how well distributed the data in a data set is. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.normal(100,10,200),np.random.normal(90,20,200))
df.plot.box(grid='True')
 
```
